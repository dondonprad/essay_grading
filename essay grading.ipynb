```python
import nltk
import string
import re, collections
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
string.punctuation
```

```python
#membersihkan teks esai menggunakan fungsi regex serta melakukan stopword dan stemming dalam bahsa indonesia
factory = StemmerFactory() #fungsi yang digunakan untuk melakukan stemmer dengan bahasa indonesia
stemmer = factory.create_stemmer()
factory_stop = StopWordRemoverFactory()
stopword = factory_stop.create_stop_word_remover() #stopword dalam bahasa indonesia 

def process_text(essay):
    essay = str(essay)
    result = re.sub('[0-9]+','', essay).lower() # menghilangkan angka dan membuat semua huruf menjadi lowercase
    result = re.sub('@[a-z0-9]+', '', result) #tanda baca akan dihilangkan 
    stem_res = stemmer.stem(result) # melakukan stemming atau mengambil kata penting dan menghilangkan imbuhan
    stop = stopword.remove(stem_res) # stopword atau menghilangkan kata yang dianggap tidak penting, sepertti kata ganti, kata hubung, dan lainnya
    return re.sub('[%s]*' % string.punctuation, '',stop) #remove punctuation
```

```python
# Jawaban mahasiswa akan dijadikan data 1 
data1 = df_essay['jawaban_mahasiswa']
# Jawaban dosen akan dijadikan data 2
data2 = df_essay['jawaban_dosen']
```

```python
data1
```

```python
data2
```

```python
df_essay['clean_jawaban_mahasiswa'] = data1.apply(process_text)
```

```python
df_essay['clean_jawaban_dosen'] = data2.apply(process_text)
```
